## Modelling

```python
df.columns
```

```python
#with present price column
df= df[['Year', 'Present_Price', 'Kms_Driven', 'Owner', 'Selling_Price',
       'Fuel_Type_Diesel', 'Fuel_Type_Petrol', 'Seller_Type_Individual',
       'Transmission_Manual']]


#without present price column
df2=df[['Year', 'Kms_Driven', 'Owner', 'Selling_Price',
       'Fuel_Type_Diesel', 'Fuel_Type_Petrol', 'Seller_Type_Individual',
       'Transmission_Manual']]
```

```python
x= df2.drop('Selling_Price', axis=1)
y=df2.Selling_Price
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor()
model.fit(x_train, y_train)
```

```python
predictions = model.predict(x_test)
```

```python
from sklearn import metrics
```

```python
import numpy as np
print('RMSE', np.sqrt(metrics.mean_squared_error(y_test, predictions)))
```

```python
import joblib
joblib.dump(model, "modeltobeused")
```

```python
feat_importances = pd.Series(model.feature_importances_, index=x_train.columns)
feat_importances.nlargest(5).plot(kind='barh', figsize=(20,5))
plt.show()
```
